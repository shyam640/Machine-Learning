### Vectors have mangintude and direction.

### Vector Operation
- Addition
- Subtraction
- Multiplication
- Angle between two vectors

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
# Plotting a vector
plt.quiver(0,0,2,4)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy', angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy', angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy', angles='xy',scale=1,color='b')
plt.quiver(0,0,-3,-6,scale_units='xy', angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# Addition of two vectors
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,3,-2])
v1 + v2
```

```python
plt.quiver(0,0,2,3,scale_units='xy', angles='xy',scale=1,color='r')
plt.quiver(0,0,3,-2,scale_units='xy', angles='xy',scale=1,color='g')
plt.quiver(0,0,5,1,scale_units='xy', angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# Subtraction of two vectors
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,3,-2])
v1 - v2
```

```python
plt.quiver(0,0,2,3,scale_units='xy', angles='xy',scale=1,color='r')
plt.quiver(0,0,3,-2,scale_units='xy', angles='xy',scale=1,color='g')
plt.quiver(0,0,-1,5,scale_units='xy', angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# Multiplying a vector by scalar
v1 = np.asarray([0,0,2,3])
2*v1
```

```python
plt.quiver(0,0,2,3,scale_units='xy', angles='xy',scale=1,color='r')
plt.quiver(0,0,4,6,scale_units='xy', angles='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

### Vector Operations
- Dot Product of 2 Vectors
- Cross Product of 2 Vectors
- Projection of Vectors

#### Dot Product

```python
a = np.array([2,3])
b = np.array([3,4])
np.dot(a,b)
```

```python
c = np.array([40,20,35])
d = np.array([34,44,72])
np.dot(c,d)
```

#### Cross Product

```python
np.cross(a,c)
```

#### Projection of a vector on b vector

```python
a = np.array([2,5])
b = np.array([8,-6])
magnitude_of_b = np.sqrt(sum(b**2))
projection_of_a_on_b = (np.dot(a,b)/magnitude_of_b**2)*b
projection_of_a_on_b
```
