0 -> Legit Transactions
1 -> Fraudulent Transactions
<br>
Above data is highly imbalanced as legit transactions are highly more than fraudulent transactions.

```python
# Seperating the legit and fraudulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
legit.shape, fraud.shape
```

## How to handle this imbalance
- Under-Sampling : It is one of the method to handle imbalance.

#### How to do it
- Build a sample dataset containing similar distribution of Legit and Fraudulent transactions.

```python
# No. of fraudulent transactions = 492
legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
# axis=0 means we want to concatenate row wise i.e. first legit_sample then fraud
# axis=1 means we want to concatenate column wise i.e. column to add after all columns
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```
