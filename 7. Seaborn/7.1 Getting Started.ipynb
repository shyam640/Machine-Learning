```python
# House Price Dataset
from sklearn.datasets import load_boston
house_dataset = load_boston()

house = pd.DataFrame(house_dataset.data, columns=house_dataset.feature_names)
house['Price'] = house_dataset.target
```

```python
house.head()
```

# Distribution Plot

```python
sns.displot(house['Price'])
```

```python
sns.distplot(house['Price'])
```

# Correlation
- Positive Correlation : In dataset if one column increase with the increase of another column then they are positively corelative.
- Negative Correlation : In dataset if one column decreases with the increase of another column then they are negatively corelative.

# Heat Map for correlation

```python
correlation = house.corr()
```

```python
# Constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='PiYG_r')


# Understanding the terms :
# Dark Color means positively Correlated and Light color means negatively correlated.
# cbar is side color bar, square is figure should be square form, fmt is no. decimal point, annot means sideways and down labelling, cmap means color of heatmap
```
