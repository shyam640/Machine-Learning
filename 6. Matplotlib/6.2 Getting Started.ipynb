### Matplotlib Example Workflow

```python
# Prepare data
from matplotlib.pyplot import title, xlabel, ylabel


x = [1,2,3,4,5,6,7]
y = [4,6,3,8,2,9,1]

# Setup plot
fig , ax = plt.subplots(figsize=(3,3))

# Plot data
ax.plot(x,y);

# Customize plot
ax.set(title='Simple Example Plot',xlabel='x-axis',ylabel='y-axis')

# Saving a plot
fig.savefig('simple_example_plot.png')
```

```python
# creating data
X = np.random.randn(10000)
fig , ax = plt.subplots()
ax.hist(X);
```

```python
# Method 1 : Drawing multiple subplots
fig , ((ax1,ax2) , (ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(X,X/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(x,y);
ax4.hist(np.random.randn(10000));
```

```python
# Method 2 : Drawing Multiple subplots
fig , ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(X,X/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(x,np.sort(y));
ax[1,1].hist(np.random.randn(100));
```

```python
# Method 3 : Creating subplots
import matplotlib.pyplot as plt
fig = plt.figure()

x = [2,4,8,9]
y = [4,2,9,7]

x1 = [2,4,8,9]
y1 = [4,2,9,7]

x2 = [2,4,8,9]
y2 = [4,2,9,7]

fig.add_subplot(2,2,1)
plt.plot(x,y,'r--')
fig.add_subplot(2,2,2)
plt.plot(x1,y1)
fig.add_subplot(2,2,3)
plt.plot(x2,y2,'y')
plt.show()
```
